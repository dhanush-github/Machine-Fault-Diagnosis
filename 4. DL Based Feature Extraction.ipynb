```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
#from keras.layers.normalization import BatchNormalization
from tensorflow import keras

from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df1=pd.read_csv('Normal_Bearing.csv')
df2=pd.read_csv('outer_race_fault_test_2.csv')
df3=pd.read_csv('outer_race_fault_test_3.csv')
df4=pd.read_csv('inner_race_fault.csv')
df5=pd.read_csv('roller_element_fault.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
df
```

```python
data = df
```

```python
encoder = LabelEncoder()
```

```python
data['Fault'] = encoder.fit_transform(data['Fault'])
```

```python
# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(data.drop('Fault', axis=1), data['Fault'], test_size=0.2, random_state=42)
```

```python
# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)
```

```python
# Build the deep learning model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=train_data.shape[1]))
#model.add(BatchNormalization())
#model.add(keras.layers.normalization.BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units=128, activation='relu'))
#model.add(BatchNormalization())
#model.add(keras.layers.normalization.BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units=256, activation='relu'))
#model.add(BatchNormalization())
#model.add(keras.layers.normalization.BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units=512, activation='relu'))
#model.add(BatchNormalization())
#model.add(keras.layers.normalization.BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(units=len(encoder.classes_), activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
model.fit(train_data, to_categorical(train_labels), validation_split=0.1, epochs=100, batch_size=32, callbacks=[early_stop])
```

```python
# Extract the deep learning features
intermediate_layer_model = Sequential()
for i in range(5):
    intermediate_layer_model.add(model.layers[i])
deep_learning_features_train = intermediate_layer_model.predict(train_data)
deep_learning_features_test = intermediate_layer_model.predict(test_data)
```
